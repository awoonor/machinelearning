## 2. Index vs Labels

A dataframe has row and column indices. Each dataframe can also have row and column labels.

```python
t.head()
```

### 2. 1 Select Rows and Columns Based on Index

```python
# t[row,column] 
# note python begins at index 0
t.iloc[1,2]
```

```python
# all columns in row with index 3
t.iloc[3]
```

```python
# above returns a series object
```

```python
# every row in column index 2
t.iloc[:,2]
```

```python
# t[rowrange, columnrange]
t.iloc[2:5,1:4]
```

### 2. 2 Select Rows and Columns Based on Index

```python
t.loc['Japan':"Uruguay", "Feb":"Apr"]
```

```python
# t [rowlist, columnlist subset of rows and columns; they don't have to be consecutive
t.loc[['France','Canada','Australia'],['Feb','Apr']]
```

```python
t1
```

```python
# rows get assigned labels if not explicitly defined
t1.loc[0:3,'Feb':'Jun']
```

## Boolean Filtering

```python
t[t['Aug']>50]
```

```python
t.loc[t['Aug']>50]
```

```python
t[(t['Aug']>50) & (t['Jul']<80)]
```
